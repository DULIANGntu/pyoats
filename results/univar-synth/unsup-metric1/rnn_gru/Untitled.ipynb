```python
import numpy as np
import jenkspy

import seaborn as sns
import matplotlib.pyplot as plt



scores = np.loadtxt("../../../../results/multivar-nasa/f1tuned-1anom/nhits/C-1-scores.txt")
scores = scores.sum(axis=1)
```

```python
sections = jenkspy.jenks_breaks(scores, nb_class=20)

fig, ax = plt.subplots(1)
sns.set_theme(style="whitegrid", palette="muted")
ax = sns.stripplot(x=scores)

for i in sections:
    ax.axvline(i)
```
